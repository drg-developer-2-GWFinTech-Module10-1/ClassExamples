```python
# Read in data
df = pd.read_csv("../Resources/liquor_sales.csv", parse_dates=True, infer_datetime_format=True)
df["datetime"] = pd.to_datetime(df["datetime"])
df.head()
df.dtypes
df = df.set_index("datetime")
df.head()
```
